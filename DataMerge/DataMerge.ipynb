```python
df_part1 = pd.merge(conflict_df, programs_df, on=['country', 'year'], how='outer', left_index=False, right_index=False)
df_part2 = pd.merge(df_part1, trade_df, on=['country', 'year'], how='outer', left_index=False, right_index=False)

df_final = pd.merge(df_part1, df_part2, on=['country', 'year', 'amount'], how='outer', left_index=False, right_index=False)
df_final = df_final.drop_duplicates()
df_final['conflict'] = False
```
