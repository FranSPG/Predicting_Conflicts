```python
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt
pd.options.display.max_rows=1000
pd.options.display.max_colwidth = 1000
```

```python
total_programs_csv = pd.read_csv('C:/Users/Franco/AnacondaProjects/Proyecto-Investigacion/total_programs.csv')
conflictos_csv = pd.read_csv('C:/Users/Franco/AnacondaProjects/Proyecto-Investigacion/conflictos.csv')
total_programs_csv.drop('Unnamed: 0', axis=1, inplace=True)
conflictos_csv.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
print(total_programs_csv.columns)
print(conflictos_csv.columns)
```

```python
final_df = pd.DataFrame(columns=['country', 'year', 'amount', 'conflict'])
```

```python
cant_country = total_programs_csv['country'].unique()
cant_years = total_programs_csv['year'].sort_values().unique()
```

```python
cant_years
```

```python
programs_country = total_programs_csv['country'].unique()
conflictos_country = conflictos_csv['location'].unique()
```

```python
country_conflictos_in_programs = conflictos_csv[conflictos_csv['location'].isin(total_programs_csv['country'])]
```

```python
for x in cant_country:
    for i in cant_years:
        try:
            amount = int(total_programs_csv.loc[(total_programs_csv['country'] == x) & (total_programs_csv['year'] == i)]['amount'])
        except:
            amount = 0

        
        final_df = final_df.append({'country': x,
                                    'year': i,
                                    'amount': amount,
                                    'conflict': 'No'}, ignore_index=True)
```

```python
cont = 0
for x in country_conflictos_in_programs.location.unique():
    for i in country_conflictos_in_programs[country_conflictos_in_programs['location'] == x]['year']:
        final_df.loc[(final_df['country'] == x ) & (final_df['year'] == i), 'conflict'] = 'Yes'
        cont = cont +1
```

```python
#print(country_conflictos_in_programs)
conflict_yes = pd.DataFrame(final_df[final_df['conflict'] == 'Yes'])
```

```python
def stats_country(country):
    df = pd.DataFrame(conflict_yes[conflict_yes['country'] == country]) #, dtype={'amount': 'int64'})
    #df = df.amount.astype('int64')
    df.reset_index(inplace=True)
    #df = df.loc[lambda df: df.amount > 0, :] SACO LAS ROWS CUANDO USA NO INVIRTIO??????
    
    max_relative_position = ((df.amount.astype(int).idxmax()+1) / len(df))
    max_amount = df['amount'].max()
    
    first_value = df.loc[lambda df: df.amount > 0, :]
    first_value.reset_index(inplace=True)
    last_value = first_value['amount'].iloc[-1]
    first_value = first_value['amount'][0]
    
    prom_amount = df['amount'].mean() # ESTAN SUMADOS LOS AMOUNT 0
    total_years_conflict = len(df)
    
    print('max: \t\t\t{}'.format(max_amount))
    print('relative position: \t{}'.format(round(max_relative_position, 4)))
    print('mean: \t\t\t{}'.format(round(prom_amount, 4)))
    print('first value: \t\t{}'.format(first_value))
    print('last value: \t\t{}'.format(last_value))
    print('total years conflict: \t{}'.format(total_years_conflict)) 
```

```python
try:
    stats_country('Afghanistan')
except:
    print('No data')
```
